```python
import sys; sys.path.insert(0, '../')
import networkx as nx, numpy as np, matplotlib.pyplot as plt
from plexsim.example import Potts
n = 32
g = nx.grid_graph([n, n])
m = Potts(g, t = 1000) 
r = m.simulate(200).reshape(-1, n, n)
print(r)
import ipywidgets as w
%matplotlib notebook
fig, ax = plt.subplots()
h = ax.imshow(r[0], vmin = 0, vmax = 1)

slider = w.IntSlider(min = 0, max = len(r) - 1)
@w.interact(t = slider)
def update(t):
    h.set_data(r[t])
    fig.canvas.flush_events()
    fig.canvas.draw()
fig.show()
```
